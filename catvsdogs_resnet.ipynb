```python
import cv2
import numpy as np
import os
import keras
from keras.applications.resnet50 import ResNet50
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Conv2D, Activation, ZeroPadding2D, Dense, MaxPooling2D, GlobalAveragePooling2D, Dropout, Input, merge, Flatten
from keras.layers.normalization import BatchNormalization
from keras.models import Model
import keras.optimizers as optimizers
```

```python
datagen = ImageDataGenerator(rotation_range=10,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
            )
```

```python
train = datagen.flow_from_directory('/input/catVSdogs/train',
                                              target_size = (224, 224),
                                              batch_size = 16,
                                              class_mode = 'categorical',
                                              
                                              )
validation = datagen.flow_from_directory('/input/catVSdogs/valid',
                                              target_size = (224, 224),
                                              batch_size = 16,
                                              class_mode = 'categorical'
                                              )
```

```python
base_model = ResNet50( weights='imagenet', include_top=False)
```

```python
base_model.summary()
```

```python
base_model.output
```

```python
#base_model.summary()
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
predictions = Dense(2, activation='softmax')(x)
new_model = Model(inputs=base_model.input, outputs= predictions)
```

```python
for layer in base_model.layers[:]:
    layer.trainable = True
```

```python
#optimizer = optimizers.Adam(lr=0.0001)
optimizer=optimizers.SGD(lr=0.0001)
new_model.compile(loss='categorical_crossentropy',
          optimizer=optimizer,
          metrics=['accuracy'])
```

```python
new_model.fit_generator(train, steps_per_epoch = 1342, validation_data=validation, validation_steps=120, epochs=3)
```
